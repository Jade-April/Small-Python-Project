# Use the textsearch API to get a list of businesses defined by the query.

```python
def get_businesses(query,api_key,min_rating=0.0,max_results=30):
    import requests,json
    api_key='AIzaSyDOStXurnpumv9_a1gdFtvl5PPKspaNg04'
    list_=[]
    
#request online data from the server
    query_=query.replace(' ','+')
    url=f'https://maps.googleapis.com/maps/api/place/textsearch/json?query={query_}&key={api_key}'
    response=requests.get(url)
    response_c=response.content
    response_=json.loads(response_c)
    
#select data that satisfies the requirement and deal with exceptions
    for item in response_['results']:
        if item['rating']>min_rating:
            name=item.get('name')
            formatted_address=item.get('formatted_address')
            opening_hours=item.get('opening_hours')
            price_level=item.get('price_level')
            rating=item.get('rating')
            list_pre=[name,formatted_address,opening_hours,price_level,rating]
            list_.append(tuple(list_pre))
            if len(list_)==max_results:
                break
    
#what if there is no appropriate answer in the first 20 results              
    while len(list_)<max_results:
        try:
            page_token=response_['next_page_token']
            new_url=f'https://maps.googleapis.com/maps/api/place/textsearch/json?pagetoken={page_token}&key={api_key}'
            response=requests.get(new_url)
            while response.json().get('status')!='OK':
                from time import sleep  
                from random import random  
                sleep(random())  
                response=requests.get(new_url)
            response_c=response.content
            response_=json.loads(response_c)
            
            for item in response_['results']:
                if item['rating']>min_rating:
                    name=item.get('name')
                    formatted_address=item.get('formatted_address')
                    opening_hours=item.get('opening_hours')
                    price_level=item.get('price_level')
                    rating=item.get('rating')
                    list_pre=[name,formatted_address,opening_hours,price_level,rating]
                    list_.append(tuple(list_pre))
                    if len(list_)==max_results:
                        break

        except KeyError:
            break 

        
    return list_
get_businesses('restaurant',api_key,max_results = 60)
```
